# Detección de Valores Nulos

Explicación: Este código analiza cada dataset para detectar valores nulos:

* Cuenta los valores nulos en cada columna
* Calcula el porcentaje que representan del total
* Imprime en pantalla un resumen con las columnas que tienen valores nulos.
* Genera un gráfico para visualizar los patrones de valores nulos encontrados.

```python
def analizar_nulos(df, titulo):
    # Calcular información sobre nulos
    total_filas = len(df)
    total_celdas = df.size
    total_nulos = df.isnull().sum().sum()

    # Calcular nulos por columna
    nulos_por_columna = df.isnull().sum()
    columnas_con_nulos = nulos_por_columna[nulos_por_columna > 0]
    porcentaje_nulos = (columnas_con_nulos / total_filas * 100).round(1)

    # Mostrar resumen textual
    print(f"=== ANÁLISIS DE VALORES NULOS: {titulo} ===")
    print(f"• Total de registros: {total_filas}")
    print(f"• Celdas con valores nulos: {total_nulos} de {total_celdas} ({(total_nulos/total_celdas*100):.2f}%)")

    if len(columnas_con_nulos) == 0:
        print(f"• No hay columnas con valores nulos")
        return

    print(f"• Columnas con valores nulos: {len(columnas_con_nulos)} de {len(df.columns)}")

    # Mostrar las columnas con mayor porcentaje de nulos
    print("\nDetalle de valores nulos por columna:")
    for col, pct in zip(porcentaje_nulos.index, porcentaje_nulos.values):
        print(f"  - {col}: {columnas_con_nulos[col]} valores ({pct}%)")

    # Crear gráfico de barras horizontal
    if len(columnas_con_nulos) > 0:
        # Ajustar el tamaño del gráfico basado en el número de columnas
        plt.figure(figsize=(12, max(4, len(columnas_con_nulos)*0.7)))

        # Crear DataFrame para el gráfico
        df_plot = pd.DataFrame({
            'columna': porcentaje_nulos.index,
            'porcentaje': porcentaje_nulos.values
        })

        # Gráfico con sintaxis actualizada y más espacio para etiquetas
        ax = sns.barplot(x='porcentaje', y='columna', data=df_plot, color='steelblue')

        # Añadir etiquetas con el valor absoluto y porcentaje
        for i, (col, pct) in enumerate(zip(columnas_con_nulos.index, porcentaje_nulos.values)):
            ax.text(min(pct + 1, 95), i, f"{columnas_con_nulos[col]} ({pct}%)", va='center')

        plt.title(f"Valores nulos en {titulo}")
        plt.xlabel("Porcentaje de registros con valor nulo (%)")
        plt.xlim(0, min(100, max(porcentaje_nulos.values) * 1.3))  # Más espacio para etiquetas

        # Usar subplots_adjust en lugar de tight_layout para mayor control
        plt.subplots_adjust(left=0.3, right=0.95, top=0.9, bottom=0.1)
        plt.show()

# Aplicar a cada dataset
analizar_nulos(sales, "Ventas")
analizar_nulos(customers, "Clientes")
analizar_nulos(products, "Productos")
analizar_nulos(stores, "Tiendas")
```

# Preparación de los Datos

```python
# Estandarizar nombres de columnas (convertir espacios a guiones bajos)
customers.columns = [col.replace(' ', '_') for col in customers.columns]
products.columns = [col.replace(' ', '_') for col in products.columns]
sales.columns = [col.replace(' ', '_') for col in sales.columns]
stores.columns = [col.replace(' ', '_') for col in stores.columns]
exchange_rates.columns = [col.replace(' ', '_') for col in exchange_rates.columns]
```

```python
# Convertir fechas
sales['Order_Date'] = pd.to_datetime(sales['Order_Date'])
sales['Delivery_Date'] = pd.to_datetime(sales['Delivery_Date'])
stores['Open_Date'] = pd.to_datetime(stores['Open_Date'])
customers['Birthday'] = pd.to_datetime(customers['Birthday'])
exchange_rates['Date'] = pd.to_datetime(exchange_rates['Date'])
```

```python
# Convertir campos de precio en Products
for col in ['Unit_Cost_USD', 'Unit_Price_USD']:  # Changed column names
    # Eliminar símbolo $ y cualquier espacio en blanco
    products[col] = products[col].str.replace('$', '', regex=False).str.strip()
    # Convertir a tipo float
    products[col] = pd.to_numeric(products[col], errors='coerce')

# Verificar conversión
print("\nCampos de precio después de conversión:")
print(products[['Unit_Cost_USD', 'Unit_Price_USD']].dtypes) # Changed column names
print(products[['Unit_Cost_USD', 'Unit_Price_USD']].head()) # Changed column names
```

# 1. Tendencia de ventas mensuales

```python
# Agregar mes a las ventas
sales['Mes_Año'] = sales['Order_Date'].dt.strftime('%Y-%m')

# Agrupar ventas por mes
ventas_mensuales = sales.groupby('Mes_Año').agg(
    total_pedidos=('Order_Number', 'nunique'),
    total_items=('Quantity', 'sum')
).reset_index()

# Visualizar tendencia
plt.figure(figsize=(12, 6))
plt.plot(ventas_mensuales['Mes_Año'], ventas_mensuales['total_pedidos'], marker='o', linewidth=2)
plt.title('Tendencia de Ventas Mensuales (2016-2021)')
plt.xlabel('Mes-Año')
plt.ylabel('Número de Pedidos')
plt.xticks(rotation=90)
plt.grid(True, alpha=0.3)
plt.tight_layout()
plt.show()
```

# Análisis de Tendencia de Ventas Mensuales (2016-2021)

## Patrones principales
- Tendencia creciente hasta finales de 2019
- Caída pronunciada en 2020-2021 (posiblemente por COVID-19)
- Patrón cíclico con caídas sistemáticas en abril de cada año

## Caídas en abril
Se observa una reducción significativa de ventas cada abril durante todo el período, con valores cercanos a mínimos anuales en 2016 (50), 2017 (≈0), 2018 (50), 2019 (75) y 2020 (80).

## Posibles causas
1. **Factores fiscales**: Fin del año fiscal en muchos países (31 de marzo)
2. **Calendario**: Semana Santa/Pascua con reducción de días laborables
3. **Operativos**: Posible período de inventario o mantenimiento programado anual

La consistencia de este patrón sugiere que es un fenómeno estructural y predecible, permitiendo desarrollar estrategias para mitigar esta caída estacional recurrente.

```python
# Calcular la fecha más reciente en los datos para referencia
fecha_maxima = sales['Order_Date'].max()

# Crear dataframe RFM
rfm = sales.groupby('CustomerKey').agg({
    'Order_Date': lambda x: (fecha_maxima - x.max()).days,  # Recency (días desde última compra)
    'Order_Number': 'nunique',  # Frequency (número de órdenes)
    'Quantity': 'sum'  # Monetary (cantidad total comprada)
})

# Renombrar columnas
rfm.columns = ['Recency', 'Frequency', 'Monetary']

# Crear cuartiles para R, F y M (4 = mejor, 1 = peor)
# Para Recency, valores menores son mejores (más recientes)
rfm['R'] = pd.qcut(rfm['Recency'], 4, labels=[4, 3, 2, 1])
# Para Frequency y Monetary, valores mayores son mejores
rfm['F'] = pd.qcut(rfm['Frequency'].rank(method='first'), 4, labels=[1, 2, 3, 4])
rfm['M'] = pd.qcut(rfm['Monetary'].rank(method='first'), 4, labels=[1, 2, 3, 4])

# Calcular puntaje RFM combinado
rfm['RFM_Score'] = rfm['R'].astype(int) + rfm['F'].astype(int) + rfm['M'].astype(int)

# Segmentar clientes según puntaje RFM
rfm['Segment'] = pd.cut(rfm['RFM_Score'],
                        bins=[0, 3, 6, 9, 12],
                        labels=['Inactivo', 'En riesgo', 'Potencial', 'Leal'])

# Mostrar resultados
print(rfm.head(10))
print("\nDistribución de segmentos:")
print(rfm['Segment'].value_counts())

# Visualizar segmentos
plt.figure(figsize=(10, 6))
sns.countplot(x='Segment', data=rfm)
plt.title('Distribución de Segmentos de Clientes')
plt.xlabel('Segmento')
plt.ylabel('Número de Clientes')
plt.show()

# Unir con información de clientes para análisis demográficos
rfm_con_info = rfm.merge(customers, on='CustomerKey', how='left')

# Visualizar segmentos por país
plt.figure(figsize=(12, 6))
segment_country = pd.crosstab(rfm_con_info['Segment'], rfm_con_info['Country'])
segment_country.plot(kind='bar', stacked=True)
plt.title('Distribución de Segmentos por País')
plt.xlabel('Segmento')
plt.ylabel('Número de Clientes')
plt.legend(title='País')
plt.xticks(rotation=0)
plt.show()
```

Implementamos un análisis RFM (Recency, Frequency, Monetary) para segmentar a los clientes basados en su comportamiento de compra. Esta técnica nos permitió categorizar a los clientes en cuatro segmentos principales:
